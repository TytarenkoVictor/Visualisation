```python
import pandas as pd
import altair as alt
```

```python
fields=['Year', 'J-D']
df = pd.read_csv('GLB.Ts_dSST.csv', skipinitialspace=True, usecols=fields)
df.Year=pd.to_datetime(df.Year, format='%Y')
df['Mean']=df['J-D'][:20].mean()
df.head()
```

```python
lst = []
for i in range(len(df['Year'])):
    if i not in [24, 64, 118, 134, 135, 136, 137]:
        lst.append('')
    else:
        lst.append(str(df['Year'][i])[:4])

a = ['' for i in range(len(df['Year']))]
a[109] = 'COLDER'
l = ['' for i in range(len(df['Year']))]
l[95] = 'HOTTER THAN THE 1880-1899 AVERAGE'

df['label'] = lst
df['a'] = a
df['l'] = l
```

```python
dot=alt.Chart(df, title="Annual Global Surface Temperature, Relative to Late 19th Century Average").mark_circle(size=60).encode(
    x=alt.X('Year', title=''),
    y=alt.Y('J-D', title=''))

mean=alt.Chart(df).mark_line(color='#040D22').encode(
    x='Year',
    y='Mean')

text = dot.mark_text(align='left', baseline='middle', dx=7).encode(text='label')

text1 = mean.mark_text(align='left', baseline='top', dx=7).encode(text='a')
text2 = mean.mark_text(align='left', baseline='bottom', dx=7).encode(text='l')

(dot + mean + text + text1+text2).properties(width=800, height=400).configure_mark(color='#e74000')
```
